# Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectFromModel
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('laptops.csv',index_col=0)
dataset.head()
```

```python
data=dataset.copy()
data['Ram(GB)']=[word.strip('GB') for word in data['Ram']]
data['Ram(GB)']=pd.to_numeric(data['Ram(GB)'])
data['Weight(kg)']=[word.strip('kg') for word in data['Weight']]
data['Weight(kg)']=pd.to_numeric(data['Weight(kg)'])
data.drop(['Weight','Ram'],axis=1,inplace=True)
data.head(50)
```

```python
numerical_features=['Inches','Ram(GB)','Weight(kg)','Price_euros']
```

```python
#for feature in numerical_features:
 #   data[feature]=np.log(data[feature])
#data.head()
```

```python
dataset=data.copy()
```

## Handling Categorical Feature


Combining Company, Product and TypeName features

```python
dataset.head(10)
```

```python
dataset['CPT']=dataset['Company']+dataset['Product']+dataset['TypeName']
dataset.drop(['Company','Product','TypeName'],axis=1,inplace=True)
```

```python
dataset.head(10)
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['Price_euros'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head(10)
```

# MODEL SELECTION AND PREDICTIONS

```python
dataset=pd.read_csv('laptops_preprocessed_v5.csv')
```

```python
y=dataset[['Price_euros']]
y.head()
```

```python
X=dataset.drop('Price_euros',axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,train_size=0.75,random_state=0)
```

```python
selected_feat=[]
for feature in dataset.columns:
    if feature not in 'Price_euros':
        selected_feat.append(feature)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model= Sequential()
model.add(Dense(128, input_dim=9, activation='relu',kernel_initializer='normal'))
model.add(Dense(256, activation='relu',kernel_initializer='normal'))
model.add(Dense(256,activation='relu',kernel_initializer='normal'))
model.add(Dense(256,activation='relu',kernel_initializer='normal'))
model.add(Dense(256,activation='linear',kernel_initializer='normal'))
#model.add(Dense(128,activation='relu',kernel_initializer='normal'))
model.add(Dense(1, activation='linear',kernel_initializer='normal'))
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100,batch_size=20,validation_data=(X_valid,y_valid))
```

```python
results = model.evaluate(X_train, y_train, batch_size=32)
print("test loss, test acc:", results)
```

```python
_, accuracy = model.evaluate(X_train, y_train,verbose=0)
print('Accuracy percentage: %.2f' % (accuracy*100))
```
